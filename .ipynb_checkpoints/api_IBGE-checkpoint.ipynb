```python
import pandas as pd
```

```python
import requests
import json

url = "https://servicodados.ibge.gov.br/api/v1/paises/BR|US"

response = requests.get(url)
json_df = response.json()

#imprime formato json leg√≠vel
print(json.dumps(response.json(), indent=3))
```

```python
df_pandas = pd.json_normalize(json_df)
```
